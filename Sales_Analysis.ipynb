# Exploratory Data Analysis of df dataframe

## Gender

```python

# plotting a bar chart for Gender and it's count from df dataframe to check whether most of the buyers are male or female

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount to check whether purchasing power of male or female is greater

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sales=sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)

for bars in sales.containers:
    sales.bar_label(bars)
```

#### From the above graphs we can see that most of the buyers are female and even the purchasing power of females is greater than male

## Age

```python
# plotting a bar chart from df dataframe to see which age group of a particular gender buys most
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

age_data=sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)

for bars in age_data.containers:
    age_data.bar_label(bars)
```

#### From the above graphs we can see that most of the buyers are females in age group of 26-35 years

## State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
state_orders=sns.barplot(data = sales_state, x = 'State',y= 'Orders')

for bars in state_orders.containers:
    state_orders.bar_label(bars)
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
state_amount=sns.barplot(data = sales_state, x = 'State',y= 'Amount')

for bars in state_amount.containers:
    state_amount.bar_label(bars)
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively


## Married Status

```python
#plotting a bar chart from df dataframe to see number of married and unmarried person
ax = sns.countplot(data = df, x = 'Married_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount/sales from married and unmarried people of a particular gender

sales_state = df.groupby(['Married_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
married_data=sns.barplot(data = sales_state, x = 'Married_Status',y= 'Amount', hue='Gender')

for bars in married_data.containers:
    married_data.bar_label(bars)
```

#### From above graphs we can see that most of the buyers are non-married females and they have high purchasing power

## Occupation

```python
#plotting a bar chart from df dataframe to see number of people from a particular occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount/sales from people of various occupations
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
occupation_data=sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')

for bars in occupation_data.containers:
    occupation_data.bar_label(bars)
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
##plotting a bar chart from df dataframe to see number of each product category bought
sns.set(rc={'figure.figsize':(24,5)})
ax = sns.countplot(data = df, x = 'Product_Category')



for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount/sales from each of the top 10 product category
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(24,5)})
product_data=sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')

for bars in product_data.containers:
    product_data.bar_label(bars)
```

#### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
#total orders from each of the top 10 Product_ID
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

#### Most of the orders is from product with Product_ID of P00265242
